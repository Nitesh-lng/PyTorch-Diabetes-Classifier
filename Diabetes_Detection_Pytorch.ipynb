```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.4,random_state=42)
```

```python
X_train_tensor=torch.from_numpy(X_train)
X_test_tensor=torch.from_numpy(X_test)
y_train_tensor = torch.tensor(y_train.values)
y_test_tensor = torch.tensor(y_test.values)
```

```python
class Custom(Dataset):
  def __init__(self,train_data,valid_data):
    self.train_data=train_data
    self.valid_data=valid_data

  def __len__(self):
    return self.train_data.shape[1]

  def __getitem__(self,index):
    return self.train_data[index] , self.valid_data[index]
```

```python
dataset=Custom(X_train_tensor,y_train_tensor)
```

```python
test_dataset=Custom(X_test_tensor,y_test_tensor)
```

```python
dataloader=DataLoader(dataset,batch_size=50,shuffle=True)
```

```python
test_dataloader=DataLoader(test_dataset,batch_size=50,shuffle=False)
```

```python
class MyNet(nn.Module):
  def __init__(self,train_len):
    super().__init__()
    self.model=nn.Sequential(
        nn.Linear(8,128),
        nn.ReLU(),
        nn.Linear(128,10),
        nn.Tanh(),
        nn.Linear(10,1),
        nn.Sigmoid()
    )

  def forward(self,data):
    return self.model(data)
```

```python
model=MyNet(X_train_tensor[1])
```

```python
epochs=100
learning_rate=1e-4
loss_fn=nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)
```

```python
for i in range(epochs):
  for batch_data,valid_data in dataloader:
    batch_data=batch_data.float()
    valid_data=valid_data.float()
    y_pred=model(batch_data)
    y_pred = y_pred.squeeze(1)
    loss=loss_fn(y_pred,valid_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  print(f'Epoch: {i + 1}, Loss: {loss.item()}')
```

```python
for i in range(epochs):
  for batch_data,valid_data in test_dataloader:
    batch_data=batch_data.float()
    valid_data=valid_data.float()
    y_pred=model(batch_data)
    y_pred = y_pred.squeeze(1)
    loss=loss_fn(y_pred,valid_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  print(f'Epoch: {i + 1}, Loss: {loss.item()}')
```

```python
model.eval()
```

```python
total = 0
correct = 0

for batch_features, batch_labels in dataloader:
  batch_features=batch_features.float()
  outputs = model(batch_features)
  _, predicted = torch.max(outputs, 1)
  total += batch_labels.size(0)
  correct += (predicted == batch_labels).sum().item()

accuracy = correct / total
print(f'Accuracy: {accuracy:.4f}')
```

```python
total = 0
correct = 0

for batch_features, batch_labels in test_dataloader:
  batch_features=batch_features.float()
  outputs = model(batch_features)
  _, predicted = torch.max(outputs, 1)
  total += batch_labels.size(0)
  correct += (predicted == batch_labels).sum().item()

accuracy = correct / total
print(f'Accuracy: {accuracy:.4f}')
```
